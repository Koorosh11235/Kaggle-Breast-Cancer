```python
# now :))))))))))))))))))))
#  this is where the magic happens

def find_best_score(model, k = 5):
    best_score = -1
    kf = KFold(n_splits=k, shuffle=True, random_state=40)
    for train_index, test_index in kf.split(x):
        x_train, x_test = x.iloc[train_index], x.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        model.fit(x_train, y_train)
        score = model.score(x_test, y_test)
        if score > best_score:
            best_score = score
            best_x_train,best_x_test = x_train,x_test
            best_y_train,best_y_test = y_train,y_test
            y_pred = model.predict(x_test)
            
            
    return best_score, best_x_train, best_x_test, best_y_train, best_y_test, y_pred
```

```python
# we get the best posssible score for the model we out inside the function
score, x_train,x_test, y_train, y_test, y_pred= find_best_score(RandomForestClassifier(), 5)
score
```
